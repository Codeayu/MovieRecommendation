```python
import pandas as pd
import numpy as np
```

```python
# Movies dataset

movies = pd.read_csv("movies.csv")
movies.head()
```

```python
# Credits dataset
credits = pd.read_csv("credits.csv")
credits.head()
```

```python
# merging the two datasets
movies = movies.merge(credits,on='title')
movies.head()
```

```python
# the features we have to add/extract from the movies dataset are: genres,movie_id,keywords,title,overview,cast,crew
movies = movies[['title','movie_id','overview','genres','keywords','cast','crew']]
movies.head()
```

```python
# checking for missing values
movies.isnull().sum()
```

```python
#remove the missing values
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
# convert the stringified features to objects
from ast import literal_eval
features = ['genres','keywords','cast','crew']
for feature in features:
    movies[feature] = movies[feature].apply(literal_eval)
```

```python
movies.dtypes
```

```python
# Convert list columns to strings to avoid TypeError
for feature in features:
	movies[feature] = movies[feature].apply(str)

# Check for duplicates
movies.duplicated().sum()
```

```python
#extracting the director from the crew feature
```

```python
movies
```

```python
#convert string to list
import ast

def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
# converting the string column to list using the existing convert function
movies["genres"] = movies["genres"].apply(convert)

```

```python
movies.head()
```

```python
#applying the convert function keywords column
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
def convert3(obj):
	l = []
	counter = 0
	for i in ast.literal_eval(obj):
		if counter != 3:
			l.append(i['name'])
			counter += 1
		else:
			break
	return l
```

```python
#extracting the cast
movies["cast"] = movies["cast"].apply(convert3)
```

```python
movies.head()
```

```python
#function to fetch the director
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
#extracting the director
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
movies.head()
```

```python
movies.dtypes
```

```python
#converting the overview column to list
movies["overview"] = movies["overview"].apply(lambda x: x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#new coloumn for tags
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
new_df['tags'] = new_df['tags'].apply(lambda x: ' '.join(x))
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
#vectorization of the tags
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(max_features=5000, stop_words='english')
vector = cv.fit_transform(new_df['tags']).toarray()
vector.shape
```

```python
cv.get_feature_names_out()
```

```python
import nltk
from nltk.stem import PorterStemmer
ps = PorterStemmer()
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
ps.stem('loving')
```

```python
# applying the stem function to the tags column
new_df['tags'] = new_df['tags'].apply(stem)
new_df.head()
```

```python
#calculating the cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(vector)
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[1]
```

```python
#function to recommend movies
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```
